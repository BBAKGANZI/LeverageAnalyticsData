```python
import pandas as pd
import seaborn as sns

df=sns.load_dataset('titanic')
```

```python
print(df.head())
```

```python
df.isna().sum()
```

```python
# age열에 나이 데이터가 없는 모든 행을 삭제 -age 열(891개중 177개의 NaN 값)
df=df.dropna(subset=['age'], how='any', axis=0)
print(len(df))
print('\n')
```

```python
# 생존이1, 사망이 0, survived
# 나이가 어릴수록 생존율이 높음. -값 나옴
df.corr()
```

```python
# embarked 열의 NaN 값을 승선도시 중에서 가장 많이 출현한 값으로 치환하기
most_freq = df['embarked'].value_counts(dropna=True).idxmax()
print(most_freq)
print('\n')
```

```python
df=df[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'embarked']]
print(df.head())
```

```python
# 원핫인코딩 - 범주형 데이터를 모형이 인식할 수 있도록 숫자형으로 변환
onehot_sex = pd.get_dummies(df['sex'])
df = pd.concat([df, onehot_sex], axis=1)
df.head()

onehot_embarked = pd.get_dummies(df['embarked'], prefix='town')
df=pd.concat([df, onehot_embarked], axis=1)

df.drop(['sex', 'embarked'], axis=1, inplace=True)
print(df.head())
```

```python
# 속성(변수) 선택
X=df[['pclass', 'age', 'sibsp', 'parch', 'female', 'male', 'town_C', 'town_Q', 'town_S']]
y=df['survived']

from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

print('train data 개수: ', X_train.shape)
print('test data 개수: ', X_test.shape)
```

```python
# sklearn 라이브러리에서 KNN 분류 모형 가져오기
from sklearn.neighbors import KNeighborsClassifier

# 모형 객체 생성(k=5로 설정)
knn = KNeighborsClassifier(n_neighbors=5)

# train data를 가지고 모형학습
knn.fit(X_train, y_train)

# test data를 가지고 y_hat을 예측(분류)
y_hat = knn.predict(X_test)

print(y_hat[0:10])
print(y_test.values[0:10])
```

```python
# 모형성능평가 - Confusion Matrix계산
from sklearn import metrics
knn_matrix = metrics.confusion_matrix(y_test, y_hat)
print(knn_matrix)
```

```python
# 모형 성능 평가 - 평가 지표 계산
knn_report = metrics.classification_report(y_test, y_hat)
print(knn_report)
```
